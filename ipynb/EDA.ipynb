### Contingency tables

```python
# Table of Overall Qual vs Kitechen Qual
kitchen_overall = pd.crosstab(index=housing_df["OverallQual"], 
                           columns=housing_df["KitchenQual"],
                             margins=True)
kitchen_overall.columns = ["Ex","Fa","Gd","TA","rowtotal"]
print ('\n Overall Qual vs Kitechen Qual')
kitchen_overall
```

```python
# Table of KitchenQual vs. ExterQual
kitchen_exter = pd.crosstab(index=housing_df["KitchenQual"], 
                           columns=housing_df["ExterQual"],
                             margins=True)
kitchen_exter.columns = ["Gd", "TA", "Ex", "Fa" ,"total"]
print ('\n KitchenQual vs. ExterQual')
kitchen_exter
```

```python
print ('\n Proportion of counts KitchenQual vs. ExterQual')
kitchen_exter.T/kitchen_exter["total"]
```

```python
# Table of Neighborhood vs. ExterQual
neighborhood_exter = pd.crosstab(index=housing_df["Neighborhood"], 
                           columns=housing_df["ExterQual"],
                             margins=True)
neighborhood_exter.columns = ["Gd", "TA", "Ex", "Fa" ,"rowtotal"]

neighborhood_exter
```

```python
neighborhood_exter.div(neighborhood_exter["rowtotal"],
                   axis=0)
```

### Distribution plot showing the distribution, mean, and median

```python
dist = sns.distplot(housing_df['SalePrice'])
dist.axvline(x=housing_df.mean()['SalePrice'], color='blue', linewidth=2, linestyle='dotted', label='mean')
dist.axvline(x=housing_df.median()['SalePrice'], color='red', linewidth=2, linestyle='dotted', label='median')
dist.legend(bbox_to_anchor=(1, 1), loc=2)
```
